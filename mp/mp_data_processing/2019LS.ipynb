```python
import pandas as pd
```

```python
data = pd.read_csv("2019ls_votes.csv")
data
```

```python
mp_data = []
const = data.constituency_name.unique()
mp_data = []
for c in const:
    
#     print(data.loc[data['constituency_name'] == c].sort_values(by=['total_votes'], ascending=False))
    temp = data.loc[data['constituency_name'] == c].sort_values(by=['total_votes'], ascending=False)
    pos = 1
    for idx, doc in temp.iterrows():
        mp = dict()
        if doc['candidate_name'].strip() == 'NOTA':
            continue
        mp['Name'] = doc['candidate_name'].strip()
        mp['Alias'] =  doc['candidate_name'].strip()
        mp['State'] = doc['province_name'].strip()
        mp['Electoral_Info'] = [{'Constituency':doc['constituency_name'].strip() , 'Party': doc['party_name'].strip() ,
                         'Position': pos, 'Year':2019}]
        pos += 1
        
        
        mp_data.append(mp)
```

```python
import pickle
with open('ls2019data','wb') as f:
    pickle.dump(mp_data, f)
```
